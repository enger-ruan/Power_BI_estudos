Introdução DAX

- No Power BI fora do Power Query .

- Vou em modelo de exibição.

- Em seguida crio uma nova coluna.

- A fonte esta sendo exibida em cima por ele que faço os comandos 'DAX'

- O primeiro parametro e o nome da nova coluna ele sera exibido dessa maneira na fonte:'  Coluna =  ' basta trocar o nome mudando a variavel 'Coluna' antes do '='

Formula DAX exemplo:

```python
Tipo do Pedido = IF('BaseVendas 2016 - 2018'[Tamanho Pedido]>1,"Pedido Múltiplo","Pedido Único")
```

Explicações desta Formula:

Tipo do Pedido  # É o nome da coluna que defini trocando de 'Coluna' para 'Tipo do Pedido'.

= IF('BaseVendas 2016 - 2018'[Tamanho Pedido]  # com o 'IF' acionado Preciso passar qual tabela e coluna quero usar na formula, a colocando entre parenteses a tabela entre aspas e a coluna em colchetes.

'  >1,"Pedido Múltiplo","Pedido Único")  #  Defino que IF(se) Tamanho Pedido for maior do que 1, 'Pedido Multiplo','Pedido unico' 

 ou seja defini que se 'Tamnho pedido' ouver um valor maior do que 1 ' Pedido Multiplo' se não ' , ' 'Pedido Unico' e como se uma variavel fosse como True e a outra como False

(lembrando que devo separar por virgula cada parametro).

> Colunas x Medidas 

- Colunas Calculadas:

* Os valores são calculados baseados na informação de cada linha da tabela(são calculdos individualmente apara cada linha, chamado de 'Contexto de Linha'.

* Acrecenta uma nova coluna na tabela

* Recalcula de acordo com as alterações nas linhas

* Geralmente usados nos campos de Linhas, Colunas, legendas ou como filtros nas Matrizes/Graficos.

Observação: Colunas calculadas geralmente são ultilizadas como filtros pros meus valores, e são aplicadas diferentemente a cada linha da tabela, por isso e criado uma coluna.

-Medidas:

* Os valores são calculados baseados nos filtros aplicados nas Matrizes chamado de 'Contexto de Filtro'.

* Não acrecenta novas colunas na tabela.

* Recalcula de acordo com alterações nos filtros do Relatório(em uma Matriz ou Griafico )

* Quase sempre usado no campo de Valores de uma Matriz/Grafico

Observação: Medidas geralmente são ultilizadas como Valores ou seja elas recebem filtros e se recalculam de acordo com oque a legenda pede(o filtro aplicado), medidas não aparecem como colunas mas como se fossem um atributo que pode ser aplicado.
